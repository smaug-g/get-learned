```python
import os
import numpy as np
import pandas as pd
```

### Keras "flow_from_generator" expects the format this program provides:
0. Place all the relevant files into the same directory as this program. 
1. Separate train_posters into "train_posters" and "validation_posters", split 200 posters 
2. Inside each folder, make 4 subfolders titled "0", "1", "2", "3"
3. Change the "train_dir" variable to the directory of your test posters, then your validation posters (one at a time)
4. run the code, it will move images into their appropriate locations --> IE img 111130 in "train" goes to /train/0 if its a poster of a genre-0 movie. 

```python
train_dir = '/Users/louissmidt/ELEC 301 Final Project/validation_posters'

info = pd.read_csv("train_data.csv", index_col=0)
meta = info[["imdbId", "Genre"]]

for filename in os.listdir(train_dir):
    if filename.endswith(".jpg"):
        row = meta.loc[meta['imdbId'] == int(filename[:-4])]
        genre = row["Genre"].values[0]
        
        os.rename(train_dir + "/" + filename, train_dir + "/" + str(genre) +"/" + filename)

    else:
        continue
print("done")        
```
