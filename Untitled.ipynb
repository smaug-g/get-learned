```python
import csv
import numpy as np
from collections import defaultdict
import sklearn
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from itertools import groupby
from operator import itemgetter
from functools import reduce
import scipy
# train_data_raw will have [(id, [metadata])]
train_data_raw = []

# all labels
y = []

# will have (id, title)
train_id_title = []

# run times
rts = []
with open('train_data.csv') as trainCSV:
    reader = csv.reader(trainCSV, delimiter=',')
    for row in reader:
        # skip the first row
        if (row[1] == 'imdbId'):
            continue
        id = int(row[0])
        train_data_raw.append((id, row[1:]))
        train_id_title.append((id, row[2]))
        y.append(int(row[6]))
        rts.append(int(row[4]))
y = np.array(y)
rts = np.array(rts)
```

```python
titles = [title[1] for title in train_id_title]
vectorizer = TfidfVectorizer()

# get data mtx
Xtfidf = np.asarray(vectorizer.fit_transform(titles).todense())
```

```python
# now, add on a dimension with the run time.
X = np.zeros([Xtfidf.shape[0], Xtfidf.shape[1] + 1])
X[:,:-1] = Xtfidf
X[:,-1] = rts
```

```python
# center data
Xc = X - np.mean(X, axis=0)

# PCA via SVD
(u, s, vh) = scipy.linalg.svd(Xc)
```

```python
# projection matrix
W = vh.T
Xc_proj = Xc.dot(W)
```
